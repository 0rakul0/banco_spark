# puxa dois arquivos ao mesmo tempo

```python
data = spark.sparkContext.textFile('./data/regiao-norte.txt,./data/regiao-sul.txt')
```

## apresenta os dados

```python
data.collect()
```

## separa por espaço, logo pega uma palavra de cada vez

```python
palavras = data.flatMap(lambda x: x.split(' '))
```

```python
import spacy
```

```python
for plv in palavras.collect():
    print(plv)
```

```python
# !pip install -U pip setuptools wheel
# !pip install -U spacy
# !python -m spacy download pt_core_news_lg
```

```python
dados_npl = spacy.load('pt_core_news_lg')
```

```python
from spacy.lang.pt.stop_words import STOP_WORDS
```

```python
for plv in palavras.collect():
    texto = dados_npl(plv)
    for t in texto:
        print(t.text, t.pos_, t.lemma_)
```

## do jeito normal

```python
def func1(x):
    minuscula = x.lower()
    return minuscula

allpalminu = palavras.map(lambda x: func1(x))
allpalminu.collect()
```

## com lambda

```python
allpalminu = palavras.map(lambda x: x.lower())
allpalminu.collect()
```

```python
# lambda -> parametro: o que tem que ser feito na função
filtro_palavras = allpalminu.filter(lambda x_palavra: x_palavra not in STOP_WORDS)
```

```python
filtro_palavras.collect()
```

# frequencia de palavras

```python
freqpal = filtro_palavras.map(lambda x: [x,1])
```

```python
freqpal.collect()
```

# junta palavras iguais

```python
from operator import add
totalpal = freqpal.reduceByKey(add)
```

```python
totalpal.collect()
```

# for interativo para colocar em ordem decendente

```python
for i in totalpal.sortBy(keyfunc=lambda x:x[1], ascending=False).collect():print(i)
```
